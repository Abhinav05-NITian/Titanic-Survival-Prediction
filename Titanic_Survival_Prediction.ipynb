```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = pd.read_csv(url)

df.head()
```

```python
print(df.shape, df.columns, df.info(), df.describe())
```

```python
df.drop("Cabin", axis=1, inplace=True)
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())    # filling missing age with median.
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])    # filling missing embarked value with mode.
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()
encoded = encoder.fit_transform(df[["Sex"]])
print(encoded.toarray())
```

```python
print(encoder.categories_)
```

```python
encoded_df = pd.DataFrame(encoded.toarray(), columns=encoder.get_feature_names_out(["Sex"]))
```

```python
encoded_df
```

```python
df = df.drop("Sex", axis=1)
```

```python
df = pd.concat([df, encoded_df], axis=1)
```

```python
df.head()
```

```python
df.drop("Name", axis=1, inplace=True)
```

```python
df
```

```python
df.drop("Ticket", axis=1, inplace=True)
df.drop("PassengerId", axis=1, inplace=True)
```

```python
df
```

```python
encoder = OneHotEncoder(sparse_output=False)
encoded = encoder.fit_transform(df[["Embarked"]])
encoded_df = pd.DataFrame(
    encoded,
    columns=encoder.get_feature_names_out(["Embarked"]),
    index=df.index
)

encoded_df.head()
```

```python
df.drop("Embarked", axis=1, inplace=True)

df = pd.concat([df, encoded_df], axis=1)
```

```python
df
```

```python
X = df.drop("Survived", axis=1)
y = df["Survived"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
print(X_train.columns)
```

```python
from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()
X_train_scaled = scalar.fit_transform(X_train)
X_test_scaled = scalar.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
tree = DecisionTreeClassifier(random_state=42)
tree.fit(X_train, y_train)
```

```python
y_pred = tree.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

```python
rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score

model = LogisticRegression(max_iter=1000)

scores = cross_val_score(
    model,
    X,
    y,
    cv=5,
    scoring="accuracy"
)

print(scores)
print(scores.mean())
```
